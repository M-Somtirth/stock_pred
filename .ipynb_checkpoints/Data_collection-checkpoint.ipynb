```python
ticker_names=pd.read_csv('Tickers.csv',index_col=False)
len(ticker_names)
```

```python
CSV_URL ="https://api.tiingo.com/tiingo/daily/ticker_name/prices?startDate=2015-01-02&token=419f55d2ac8913929c3054133171783b621cdc06"

for i, r in ticker_names.iterrows():
    CSV_URL_temp = CSV_URL.replace('ticker_name', str(r['TICKER']))
    with requests.Session() as s:
        try:
            download = s.get(CSV_URL_temp)
            decoded_content = download.content.decode('utf-8')
            df = pd.read_json(io.StringIO(decoded_content))
            df.to_csv(str(r['COMPANY NAME']) + '_data.csv', index=False)
        except ValueError as e:
            print(f"ValueError: {e} for {r['TICKER']}")
        except Exception as e:
            print(f"An error occurred: {e} for {r['TICKER']}")

```

```python
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df_pearl=pd.read_csv('QUALCOMM Incorporated_data.csv',index_col='date',parse_dates=True)
df_pearl.head()
```

```python
df_pearl['high'].plot(figsize=(12,6))
```

```python
df_pearl.index = pd.to_datetime(df_pearl.index)

df_pearl = df_pearl.asfreq('D')
```

```python
res=seasonal_decompose(df_pearl['high'],period=365)
res.plot();

# res.seasonal.plot(figsize=(16,8))
```

```python
df_pearl['high'].fillna(method='ffill', inplace=True)
```

```python
auto_arima(df_pearl['high'], m=7, trace=True).summary()
```

```python
train_df=df_pearl[:1100]
test_df=df_pearl[1100:]
model=SARIMAX(df_pearl['high'], order=(4,1,3),seasonal_order=(1,0,1,7))
res=model.fit()
start=len(train_df)
end= len(train_df)+len(test_df)-1
pred=res.predict(start,end).rename('Prediction')
ax=test_df['high'].plot(legend=True, figsize=(14,8))
pred.plot(legend=True)
```

```python
from statsmodels.tools.eval_measures import rmse
rmse(test_df['high'],pred)
```
